```python
df=df.drop(columns={'Patient ID','Country',
       'Continent', 'Hemisphere','Income'})
```

```python
df[['Systolic','Dystolic']]=df['Blood Pressure'].str.split('/',expand=True).astype(int)
df=df.drop(columns={'Blood Pressure'})
```

```python
catvar=['Sex','Diet']
df2 = pd.get_dummies(df, columns=catvar, drop_first=True)
```

```python
x=df2.drop(['Heart Attack Risk'], axis=1)
y=df2['Heart Attack Risk']
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros=RandomOverSampler(sampling_strategy='not majority')
x,y=ros.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, log_loss, roc_auc_score, classification_report
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
lr = LogisticRegression()

svm = SVC()

dtc =DecisionTreeClassifier()

knc = KNeighborsClassifier()
```

```python
lr.fit(X_train, y_train)
svm.fit(X_train, y_train)
dtc.fit(X_train, y_train)
knc.fit(X_train, y_train)
```

```python
score_lr = lr.score(X_test, y_test)
score_svm = svm.score(X_test, y_test)
score_dtc = dtc.score(X_test, y_test)
score_knc = knc.score(X_test, y_test)

modelscores={
    'Model': ['LogisticRegression', 'SupportVectorMachine', 'DecisionTreeClassifier', 'KNeighborsClassifier'],
    'Score': [score_lr, score_svm, score_dtc, score_knc]
}
```

```python
modelscoresdf=pd.DataFrame(modelscores)
modelscoresdf
```

```python
conf_mtx = []
for i, model_name in zip([lr, svm, dtc, knc],
                                ['LogisticRegression', 'SupportVectorMachine', 'DecisionTreeClassifier', 'KNeighborsClassifier']):

    y_pred = i.predict(X_test)
    accuracy = confusion_matrix(y_test, y_pred)
    conf_mtx.append((model_name, accuracy))

conf_mtx_df = pd.DataFrame(conf_mtx, columns=['Model', 'Confusion Matrix'])
```

```python
conf_mtx_df
```

```python
lr.fit(X_train, y_train)
y_lr = lr.predict(X_test)
print(classification_report(y_lr,y_test))
```

```python
y_pred_proba_ll = lr.predict_proba(X_test)[:,1]
loss = log_loss(y_test, y_pred_proba_ll)
loss
print(f' The Log Loss for the Logistic Regression model is {loss}')
```

```python
y_pred_proba = lr.predict_proba(X_test)[:,1]
auc_roc = roc_auc_score(y_test, y_pred_proba)
print(f' The Area under the ROC Curve for the Logistic Regression model is is {auc_roc}')
```

```python
from sklearn.model_selection import cross_val_score
```

```python
acc = cross_val_score(estimator = lr, X=X_train, y=y_train, cv=10)
```

```python
print(f' The cross validation scores are {acc}')
```

```python
print(f' The mean cross validation score is {acc.mean()}')
print(f' The standard deviation for the cross validation score is {acc.std()}')
```
